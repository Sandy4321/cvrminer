```python
from cvrminer.xbrler import Regnskabsdata1000
```

```python
rd = Regnskabsdata1000()
```

```python
tag_matrix = DataFrame(rd.tag_matrix()).T.fillna(0).to_sparse()
```

```python
tag_matrix.head()
```

```python
binarized_tag_matrix = (tag_matrix.to_dense() == 0).astype(int).to_sparse()
```

```python
model = NMF(n_components=15)
```

```python
W, H = model.fit_transform(binarized_tag_matrix), model.components_
```

```python
n_items = 20
for n in range(H.shape[0]):
    index = argsort(-H[n, :])
    print(n)
    print("\n".join(["{} {}".format(v, c) for v, c in zip(H[n, index[:n_items]],tag_matrix.columns[index[:n_items]].values)]))
    print('---'* 10)
```
